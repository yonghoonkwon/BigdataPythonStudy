```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 14, 100)
y1 = np.sin(x+0.0)*1
y2 = np.sin(x+0.5)*2
y3 = np.sin(x+1.0)*3
y4 = np.sin(x+2.0)*4

#sns.set_style('white')
#sns.set_style('dark')
sns.set_style('whitegrid')

plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)

plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
sns.boxplot(x=tips['total_bill'])
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker',data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker', palette='Set3', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', hue='smoker', color='0.5', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, size=7)
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', palette='Set1', data=tips, size=7)
```

```python
uniform_data = np.random.rand(10, 12)
uniform_data
sns.heatmap(uniform_data, annot=True)
```

```python
flights = sns.load_dataset('flights')
flights.head(5)

flights = flights.pivot('month', 'year', 'passengers')
flights.head(5)

plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True)
plt.show()
```

```python
sns.set(style='ticks')
iris = sns.load_dataset('iris')
iris.head(10)

sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue='species')
plt.show()
sns.pairplot(iris, vars=['sepal_width', 'sepal_length'])
plt.show()
```

```python
sns.pairplot(iris, x_vars=['sepal_width', 'sepal_length'], y_vars=['petal_width', 'petal_length'])
plt.show()
```
